## 1. Introduction

In this assignment, I apply Linear Regression techniques to three different datasets:

## 2. Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
```

## 3. Height and Weight Dataset

### 3.1. Load and Explore the Data
First, we load the dataset and take a quick look at its structure and sample records to understand the data.

```python
# Load dataset
df_height_weight = pd.read_csv('new_height_weight.csv')
df_height_weight.head()
```

### 3.2. Split Data into Train and Test Sets
We divide the dataset into training and test sets. The training set will be used to fit the model, while the test set will be used to evaluate its performance.

```python
X_hw = df_height_weight[['Height']]
y_hw = df_height_weight['Weight']
X_train_hw, X_test_hw, y_train_hw, y_test_hw = train_test_split(X_hw, y_hw, test_size=0.2, random_state=42)
```

### 3.3. Train Linear Regression Model
We create a Linear Regression model and train it using the training data.

```python
model_hw = LinearRegression()
model_hw.fit(X_train_hw, y_train_hw)
```

### 3.4. Make Predictions and Evaluate
The trained model is used to make predictions on the test data. We then evaluate the model's accuracy using metrics such as Mean Absolute Error (MAE) and R² Score.

```python
y_pred_hw = model_hw.predict(X_test_hw)
mae_hw = mean_absolute_error(y_test_hw, y_pred_hw)
r2_hw = r2_score(y_test_hw, y_pred_hw)

print(f"Mean Absolute Error: {mae_hw:.2f}")
print(f"R^2 Score: {r2_hw:.2f}")
```

### 3.5. Visualizing the Results

We visualize the actual vs. predicted values to better understand how well the model performs and to identify any patterns or errors.

```python
# Visualization
plt.scatter(X_test_hw, y_test_hw, color='blue', label='Actual')
plt.scatter(X_test_hw, y_pred_hw, color='red', alpha=0.5, label='Predicted')
plt.title("Height vs. Weight - Actual vs Predicted")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.legend()
plt.show()
```


**Evaluation:**  
The predicted values (red) closely follow the actual data (blue), indicating that the model fits the data very well and the relationship between height and weight is strongly linear.

---

Overall, the linear regression model achieved high accuracy on this dataset. This is expected, as the relationship between height and weight is almost perfectly linear.

## 4. Human Brain Weight and Head Size Dataset

### 4.1. Load and Explore the Data

```python
df_brain = pd.read_csv('HumanBrain_WeightandHead_size.csv')
df_brain.head()
```

- **Feature:** Head Size (cm³)
- **Label:** Brain Weight (grams)

### 4.2. Split Data into Train and Test Sets

```python
X_brain = df_brain[['Head Size(cm^3)']]
y_brain = df_brain['Brain Weight(grams)']
X_train_brain, X_test_brain, y_train_brain, y_test_brain = train_test_split(X_brain, y_brain, test_size=0.2, random_state=42)
```

### 4.3. Train Linear Regression Model

```python
model_brain = LinearRegression()
model_brain.fit(X_train_brain, y_train_brain)
```

### 4.4. Make Predictions and Evaluate

```python
y_pred_brain = model_brain.predict(X_test_brain)
mae_brain = mean_absolute_error(y_test_brain, y_pred_brain)
r2_brain = r2_score(y_test_brain, y_pred_brain)

print(f"Mean Absolute Error: {mae_brain:.2f}")
print(f"R^2 Score: {r2_brain:.2f}")
```

```python
plt.scatter(X_test_brain, y_test_brain, color='blue', label='Actual')
plt.scatter(X_test_brain, y_pred_brain, color='red', alpha=0.5, label='Predicted')
plt.title("Head Size vs. Brain Weight - Actual vs Predicted")
plt.xlabel("Head Size (cm³)")
plt.ylabel("Brain Weight (grams)")
plt.legend()
plt.show()
```

**Evaluation:**  
The predicted brain weights generally align with the actual values, though there are some deviations at the extremes. The linear model captures most of the trend but not all variations.

---
The model performed well, explaining 71% of the variance in brain weight. However, outliers and biological variability mean a perfect fit is less likely.

## 5. Boston Housing Dataset

### 5.1. Load and Explore the Data

```python
import pandas as pd

# Sütun isimlerini baştan tanımla
column_names = [
    'CRIM','ZN','INDUS','CHAS','NOX','RM','AGE',
    'DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV'
]

# Veriyi başlıklarla oku
df_boston = pd.read_csv('boston_housing.csv', names=column_names, header=None)
df_boston.head()
```

- **Features:** All columns except `MEDV` (target)
- **Label:** `MEDV` (Median value of owner-occupied homes in $1000's)

```python
# Eksik veri içeren satırları kaldır
df_boston = df_boston.dropna()
```

### 5.2. Split Data into Train and Test Sets

```python
X_boston = df_boston.drop('MEDV', axis=1)
y_boston = df_boston['MEDV']
X_train_boston, X_test_boston, y_train_boston, y_test_boston = train_test_split(X_boston, y_boston, test_size=0.2, random_state=42)
```

### 5.3. Train Linear Regression Model

```python
model_boston = LinearRegression()
model_boston.fit(X_train_boston, y_train_boston)
```

### 5.4. Make Predictions and Evaluate

```python
y_pred_boston = model_boston.predict(X_test_boston)
mae_boston = mean_absolute_error(y_test_boston, y_pred_boston)
r2_boston = r2_score(y_test_boston, y_pred_boston)

print(f"Mean Absolute Error: {mae_boston:.2f}")
print(f"R^2 Score: {r2_boston:.2f}")
```

```python
plt.scatter(y_test_boston, y_pred_boston, alpha=0.7)
plt.xlabel("Actual MEDV")
plt.ylabel("Predicted MEDV")
plt.title("Actual vs Predicted - Boston Housing")
plt.show()
```

**Evaluation:**  
Predicted house prices follow the general trend of actual values, but the scatter indicates some non-linearity or additional factors affecting housing prices. The model explains a significant portion, but not all, of the variability.

---

Linear regression provides a reasonable estimate for house prices, but the relatively lower R² suggests more complex models or additional features may improve performance.
